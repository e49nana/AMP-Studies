## Produit scalaire et angle
$\langle u, v \rangle = \|u\|\|v\|\cos\theta$

```python
from dot_product import *
import numpy as np
u = np.array([1, 2, 3])
v = np.array([4, -1, 2])
print(f'⟨u,v⟩ = {np.dot(u, v)}')
print(f'angle = {np.degrees(np.arccos(np.dot(u,v)/(np.linalg.norm(u)*np.linalg.norm(v)))):.1f}°')
```
