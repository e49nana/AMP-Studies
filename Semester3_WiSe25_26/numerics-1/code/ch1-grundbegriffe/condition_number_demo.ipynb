# Conditionnement de problèmes mathématiques

**Référence :** Kröger, §1.2.3–1.2.5, Satz 1.9, Beispiel 1.10.

## Définition (Satz 1.9)

Pour $f : \mathbb{R} \to \mathbb{R}$ différentiable :

$$\text{cond}_{abs}(f, x) = |f'(x)|, \qquad \text{cond}_{rel}(f, x) = \frac{|f'(x)| \cdot |x|}{|f(x)|}$$

Le conditionnement relatif est le **facteur d'amplification** des erreurs relatives :
$$\frac{|\Delta f|}{|f|} \leq \text{cond}_{rel} \cdot \frac{|\Delta x|}{|x|}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from condition_number import (
    condition_relative, condition_sqrt, condition_exp,
    condition_log, condition_sin, condition_soustraction,
    tracer_condition, tracer_amplification,
)
```

## 1. Exemples canoniques (Beispiel 1.10)

| Fonction | $\text{cond}_{rel}$ | Commentaire |
|---|---|---|
| $\sqrt{x}$ | $1/2$ | toujours bien conditionné |
| $e^x$ | $|x|$ | mal conditionné pour $|x| \gg 1$ |
| $\ln x$ | $1/|\ln x|$ | mal conditionné pour $x \approx 1$ |
| $\sin x$ | $|x \cot x|$ | mal conditionné aux multiples de $\pi$ |

```python
for name, c in [
    ('√x,  x=4', condition_sqrt(4)),
    ('eˣ,  x=1', condition_exp(1)),
    ('eˣ,  x=50', condition_exp(50)),
    ('ln x, x=2', condition_log(2)),
    ('ln x, x=1.001', condition_log(1.001)),
    ('sin x, x=π/4', condition_sin(np.pi/4)),
]:
    print(f'{name:20s} cond_rel = {c:.4f}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
tracer_condition(ax=axes[0])
tracer_amplification(ax=axes[1])
plt.tight_layout()
plt.show()
```

## 2. Soustraction : le cas catastrophique

$$\text{cond}_{rel}(x_1 - x_2) = \frac{|x_1| + |x_2|}{|x_1 - x_2|}$$

Quand $x_2 \to x_1$, le conditionnement **explose** — c'est la source fondamentale de l'Auslöschung (programme 2).

```python
for x2 in [0.0, 0.9, 0.999, 0.999999, 0.999999999]:
    print(f'1 - {x2} : cond = {condition_soustraction(1.0, x2):.2e}')
```
