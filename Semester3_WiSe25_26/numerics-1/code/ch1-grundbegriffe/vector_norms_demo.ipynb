## 1. Définition (rappel)

Une application $\|\cdot\| : V \to \mathbb{R}$ est une **norme** si :

| Axiome | Formule |
|---|---|
| Positivité | $\|x\| \geq 0$ |
| Définitude | $\|x\| > 0$ pour $x \neq 0$ |
| Homogénéité positive | $\|\lambda x\| = \|\lambda\| \, \|x\|$ |
| Inégalité triangulaire | $\|x + y\| \leq \|x\| + \|y\|$ |

Les **p-normes** sur $\mathbb{R}^n$ :
$$\|x\|_p = \left( \sum_{i=1}^n |x_i|^p \right)^{1/p}, \quad p \geq 1$$
$$\|x\|_\infty = \max_{i=1,\dots,n} |x_i|$$

## 3. Vérification numérique des axiomes

On échantillonne aléatoirement des vecteurs et on vérifie que les 4 axiomes tiennent.

**Important :** ce n'est pas une preuve mathématique — c'est un *sanity check*. Si un axiome échoue sur un échantillon aléatoire, la fonction n'est pas une norme. Si tous passent, c'est un indice fort.

```python
for nom, norme in [("||·||_1", norm_1), ("||·||_2", norm_2), ("||·||_∞", norm_inf)]:
    print(f"--- {nom} ---")
    print(verifier_axiomes(norme, n=10))
```

### Contre-exemple : ce qui *n'est pas* une norme

Pour $0 < p < 1$, la quantité $\left(\sum |x_i|^p\right)^{1/p}$ ne vérifie **pas** l'inégalité triangulaire — ce n'est pas une norme.

```python
fausse_norme = lambda x: float(np.sum(np.abs(x) ** 0.5) ** 2)
rapport = verifier_axiomes(fausse_norme, n=10)
print(rapport)
print(f"\nTous les axiomes passent ? {rapport.tous_verifies}")
```

## 4. Boules unitaires (Übung 1.3)

Les ensembles $\{x \in \mathbb{R}^2 : \|x\|_p \leq 1\}$ ont des formes très différentes selon $p$ :

- $p = 1$ : losange (carré tourné de 45°)
- $p = 2$ : disque
- $p = \infty$ : carré

Pour $1 < p < 2$ on a une forme intermédiaire entre losange et disque ; pour $p > 2$ entre disque et carré.

```python
tracer_boules_unitaires(valeurs_p=(1, 1.5, 2, 4, np.inf))
plt.show()
```

## 5. Équivalence des normes (Satz 1.5, Beispiel 1.6)

**Théorème :** en dimension finie, toutes les normes sont équivalentes : pour toutes normes $\|\cdot\|_a, \|\cdot\|_b$ il existe $L, M > 0$ tels que
$$\|x\|_a \leq L \cdot \|x\|_b \quad \text{et} \quad \|x\|_b \leq M \cdot \|x\|_a.$$

**Beispiel 1.6 :** pour $\|\cdot\|_1$ et $\|\cdot\|_\infty$ on a
$$\|x\|_\infty \leq \|x\|_1 \leq n \cdot \|x\|_\infty.$$

Vérifions empiriquement que les constantes optimales sont bien $L = 1$ et $M = n$.

```python
print(f"{'n':>4} | {'L_emp':>10} | {'L_théo':>8} | {'M_emp':>10} | {'M_théo':>8}")
print("-" * 50)
for n in [2, 5, 10, 50, 100]:
    L, M = constantes_equivalence_empirique(norm_inf, norm_1, n=n, n_essais=20_000)
    print(f"{n:>4} | {L:>10.4f} | {1:>8} | {M:>10.4f} | {n:>8}")
```

**Observations :**

- $L_\text{emp} \to 1$ par valeurs inférieures (la borne $\|x\|_\infty \leq \|x\|_1$ est atteinte quand un seul $x_i$ est non nul, ce qui est rare en tirage gaussien).
- $M_\text{emp} \to n$ par valeurs inférieures (la borne $\|x\|_1 \leq n \cdot \|x\|_\infty$ est atteinte quand toutes les composantes ont la même valeur absolue).

**Attention pratique :** quand $n$ est grand, $M = n$ explose. L'équivalence théorique cache le fait que numériquement, choisir la bonne norme peut changer beaucoup de choses.
