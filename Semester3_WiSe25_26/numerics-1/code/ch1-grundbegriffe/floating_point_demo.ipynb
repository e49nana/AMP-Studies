## 1. Propriétés du `float64`

Le standard IEEE 754 double precision utilise 8 octets (64 bits) :
- 1 bit de signe
- 11 bits d'exposant (biais 1023)
- 52 bits de mantisse

$$x = (-1)^s \cdot (1.m_{51}m_{50}\dots m_0)_2 \cdot 2^{e - 1023}$$

Maschinengenauigkeit : $\varepsilon_{mach} = 2^{-52} \approx 1{,}11 \cdot 10^{-16}$.

```python
import numpy as np
from floating_point import info_ieee754, epsilon_machine_experimental

for k, v in info_ieee754().items():
    print(f'{k:25s} : {v}')
print(f'\nε_mach expérimental = {epsilon_machine_experimental():.6e}')
```

## 2. Représentation binaire

Le nombre $0.1$ n'est **pas** représentable exactement en binaire :
$$0.1_{10} = 0.0\overline{0011}_{2}$$

```python
from floating_point import analyser_float

for x in [1.0, 0.1, -3.14]:
    info = analyser_float(x)
    print(f'{x:>8} : {info["bits"]}')
    print(f'           = {info["reconstruction"]}')
```

## 3. Non-associativité

En arithmétique exacte : $(a + b) + c = a + (b + c)$.
En flottants : **faux** en général.

```python
from floating_point import demo_non_associativite
for label, val in demo_non_associativite():
    print(f'{label:25s} = {val}')
```

## 4. Somme de Kahan

L'erreur de la somme naïve croît en $O(n \cdot \varepsilon_{mach})$.
L'algorithme de Kahan la maintient en $O(\varepsilon_{mach})$ indépendamment de $n$.

```python
import matplotlib.pyplot as plt
from floating_point import tracer_erreur_sommation
tracer_erreur_sommation()
plt.show()
```

## 5. Condition des opérations (§1.3.3)

| Opération | $\text{cond}_{rel,\infty}$ |
|---|---|
| $x_1 \cdot x_2$ | $2$ (toujours bien conditionné) |
| $x_1 / x_2$ | $2$ |
| $x_1 + x_2$ | $\frac{|x_1|+|x_2|}{|x_1+x_2|}$ (Auslöschung si $x_2 \approx -x_1$) |

```python
from floating_point import condition_operation
for op, x1, x2 in [('+',1,2), ('+',1,-0.999), ('-',1.001,1.0), ('*',3,7)]:
    c = condition_operation(op, x1, x2)
    print(f"{x1} {op} {x2} : cond = {c['cond_rel_∞']:.2f}")
```
