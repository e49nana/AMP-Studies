# Programme 8 — Newton pour systèmes non-linéaires

**Référence :** Kröger, *Numerische Mathematik 1 für AMP*, section 3.2.

1. Newton multivarié : $x^+ = x - J(x)^{-1} f(x)$ via résolution de $Js = -f(x)$
2. Jacobienne analytique vs différences finies
3. Reproduction de l'**Übung 3.10**
4. Convergence quadratique vérifiée expérimentalement

```python
import numpy as np
import matplotlib.pyplot as plt
from newton_systems import (
    newton_systeme, newton_systeme_df, jacobienne_df,
    f_uebung_3_10, jac_uebung_3_10, f_rosenbrock, jac_rosenbrock,
    ordre_experimental,
    tracer_convergence_systeme, tracer_trajectoire_2d,
)
```

## 1. L'algorithme (section 3.2.3)

À chaque pas, on résout le **système linéaire** $J(x) \, s = -f(x)$ (formule 3.13), puis $x^+ = x + s$.

**Point crucial du script :** on ne calcule **jamais** $J^{-1}$ explicitement. On résout le système linéaire, par exemple via la décomposition LR du programme 3.

## 2. Übung 3.10

$$\begin{cases} 2(x_1+x_2)^2 + (x_1-x_2)^2 = 8 \\ 5x_1^2 + (x_2-3)^2 = 9 \end{cases}, \quad x_0 = (2, 0)^T$$

```python
x0 = np.array([2.0, 0.0])
res_a = newton_systeme(f_uebung_3_10, jac_uebung_3_10, x0)
res_df = newton_systeme_df(f_uebung_3_10, x0)

print(f'Jac. analytique   : x* = {res_a.x}, {res_a.iterations} itérations')
print(f'Jac. diff. finies : x* = {res_df.x}, {res_df.iterations} itérations')
print(f'\nOrdres expérimentaux : {[f"{o:.3f}" for o in ordre_experimental(res_a.historique_x, res_a.x)]}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
tracer_trajectoire_2d(f_uebung_3_10, [res_a], xlim=(-1, 3), ylim=(-1, 4), ax=axes[0])
axes[0].set_title('Übung 3.10 — trajectoire')
tracer_convergence_systeme([res_a, res_df], ax=axes[1])
plt.tight_layout()
plt.show()
```

Le graphique de gauche montre les deux courbes $f_1 = 0$ (bleu) et $f_2 = 0$ (rouge). La solution $x^* = (1, 1)^T$ est à leur intersection. La trajectoire de Newton y arrive en 5 pas.

## 3. Rosenbrock — convergence en 2 pas

```python
res_ros = newton_systeme(f_rosenbrock, jac_rosenbrock, np.array([-1.0, 1.0]))
print(res_ros)
print(f'x* = {res_ros.x}')
```
