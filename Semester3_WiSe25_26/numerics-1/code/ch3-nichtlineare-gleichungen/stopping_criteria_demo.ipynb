# Critères d'arrêt — Comparaison

**Référence :** Kröger, §3.1.5.

## Les 3 critères

| # | Critère | Quand il se déclenche |
|---|---|---|
| $\tau_1$ | $|f(x^+)| \leq \tau_1$ | résidu petit |
| $\tau_2$ | $|x^+ - x| / |x^+| \leq \tau_2$ | changement relatif petit |
| $\tau_3$ | $|x^+ - x| \leq \tau_3$ | changement absolu petit |

**Piège classique :** un résidu petit **ne garantit pas** une solution précise
(et inversement). Exemple : $f(x) = (x-1)^3$ a un résidu de $10^{-6}$ en $x = 0.99$
mais l'erreur est de 1%.

## Newton avec les 3 critères tracés

```python
f = lambda x: x**2 - 3
df = lambda x: 2*x
trace = newton_avec_criteres(f, df, 4.5, tau1=1e-10, tau2=1e-8, tau3=1e-14)
print(f'τ₁ (résidu)  : itération {trace.arret_residu}')
print(f'τ₂ (relatif) : itération {trace.arret_relatif}')
print(f'τ₃ (absolu)  : itération {trace.arret_absolu}')
```

```python
tracer_criteres([trace], np.sqrt(3))
plt.show()
```
