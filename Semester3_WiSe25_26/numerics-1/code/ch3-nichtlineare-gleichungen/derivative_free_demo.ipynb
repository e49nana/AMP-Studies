## 1. Le problème de Regula Falsi

Regula Falsi remplace la bissection par une interpolation linéaire :
$$x_{k+1} = a - f(a) \cdot \frac{b - a}{f(b) - f(a)}$$

**Défaut :** un des bords peut rester « collé » (*stuck endpoint*) → convergence dégénère en linéaire lente.

```python
import numpy as np
import matplotlib.pyplot as plt
from derivative_free import regula_falsi, illinois, hybride_bisection_secante, tracer_comparaison

f = lambda x: x**3 - 2*x - 5
x_star = 2.0945514815423265

res_rf = regula_falsi(f, 2.0, 3.0)
res_il = illinois(f, 2.0, 3.0)
res_hy = hybride_bisection_secante(f, 2.0, 3.0, n_bisect=5)

for r in [res_rf, res_il, res_hy]:
    print(r)
```

## 2. Illinois : débloquer le bord collé

L'astuce de l'Illinois : quand un bord ne bouge pas, diviser sa valeur $f$ par 2.
Résultat : convergence super-linéaire ($\alpha \approx 1.442$).

## 3. Stratégie hybride (§3.1.6)

1. Quelques pas de **bissection** (convergence globale garantie)
2. Basculer vers la **sécante** ($\alpha \approx 1.618$)

Combine le meilleur des deux : robustesse + vitesse.

```python
tracer_comparaison([res_rf, res_il, res_hy], x_star)
plt.show()
```
