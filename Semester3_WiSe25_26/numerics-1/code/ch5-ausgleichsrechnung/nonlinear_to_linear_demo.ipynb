# Linéarisation de modèles non-linéaires

**Référence :** Kröger, §5.5.3.

## Principe

Transformer un modèle non-linéaire en un problème linéaire par changement de variable :

| Modèle | Transformation | Problème linéaire |
|---|---|---|
| $y = a \cdot e^{bx}$ | $\ln y = \ln a + bx$ | linéaire en $(\ln a, b)$ |
| $y = a \cdot x^b$ | $\ln y = \ln a + b \ln x$ | log-log |
| $y = a/(b+x)$ | $1/y = b/a + x/a$ | inverse |
| $y = ax/(b+x)$ | $1/y = 1/a + (b/a)/x$ | Michaelis-Menten |
| $y = a + b \ln x$ | déjà linéaire | en $(1, \ln x)$ |

```python
import numpy as np
import matplotlib.pyplot as plt
from nonlinear_to_linear import (
    fit_exponentiel, fit_puissance, fit_saturation,
    fit_logarithmique, tracer_linearisation,
)
```

## 1. Croissance exponentielle : $y = 3 e^{0.5x}$

```python
rng = np.random.default_rng(42)
x1 = np.linspace(0, 4, 25)
y1 = 3.0 * np.exp(0.5*x1) + rng.normal(0, 0.3, 25)
y1 = np.maximum(y1, 0.1)
m1 = fit_exponentiel(x1, y1)
print(f'a = {m1.a:.4f} (vrai 3), b = {m1.b:.4f} (vrai 0.5), R² = {m1.r_squared:.4f}')
```

## 2. Loi de puissance : $y = 2 x^{1.5}$

```python
x2 = np.linspace(0.5, 10, 30)
y2 = 2.0 * x2**1.5 + rng.normal(0, 1, 30)
y2 = np.maximum(y2, 0.1)
m2 = fit_puissance(x2, y2)
print(f'a = {m2.a:.4f} (vrai 2), b = {m2.b:.4f} (vrai 1.5), R² = {m2.r_squared:.4f}')
```

## 3. Saturation (Michaelis-Menten) : $y = 10x/(5+x)$

```python
x3 = np.linspace(0.5, 20, 20)
y3 = 10*x3/(5+x3) + rng.normal(0, 0.3, 20)
y3 = np.maximum(y3, 0.1)
m3 = fit_saturation(x3, y3)
print(f'a = {m3.a:.4f} (vrai 10), b = {m3.b:.4f} (vrai 5), R² = {m3.r_squared:.4f}')
```

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
tracer_linearisation(x1, y1, m1, axes[0,0], axes[0,1])
tracer_linearisation(x2, y2, m2, axes[1,0], axes[1,1])
tracer_linearisation(x3, y3, m3, axes[2,0], axes[2,1])
plt.tight_layout()
plt.show()
```

**Gauche :** espace original (données + courbe ajustée).
**Droite :** espace linéarisé (les points doivent s'aligner).

Si les points ne s'alignent pas dans l'espace linéarisé, le modèle est inadapté.
