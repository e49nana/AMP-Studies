# Décomposition QR par Householder

**Référence :** Kröger, §5.3.

## Réflexion de Householder (Satz 5.6)

Pour un vecteur $a$, on construit $w = a + \text{sign}(a_1)\|a\|_2 e_1$, puis
$$Q = I - \frac{2 w w^T}{w^T w}$$

envoie $a$ sur $\mp\|a\|_2 e_1$. Appliquée colonne par colonne, on obtient $QA = R$.

```python
import numpy as np
import matplotlib.pyplot as plt
from householder_qr import (
    decomposition_qr, moindres_carres_qr, moindres_carres_normales,
    tracer_ajustement, tracer_stabilite_comparison,
)
```

## 1. Vérification QR

```python
A = np.array([[1,1,1],[1,2,4],[1,3,9],[1,4,16]], dtype=float)
qr = decomposition_qr(A)
print(f'||QᵀQ - I|| = {np.linalg.norm(qr.Q.T @ qr.Q - np.eye(4), np.inf):.2e}')
print(f'||QA - R||  = {np.linalg.norm(qr.Q @ A - qr.R, np.inf):.2e}')
```

## 2. QR vs équations normales (§5.4)

$\kappa(A^T A) = \kappa(A)^2$ → les équations normales perdent le **double** de chiffres.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
rng = np.random.default_rng(42)
x = np.linspace(0, 5, 30)
y = 2.5*x + 1 + rng.normal(0, 0.5, 30)
tracer_ajustement(x, y, degres=(1, 3, 8), ax=axes[0])
tracer_stabilite_comparison(ax=axes[1])
plt.tight_layout()
plt.show()
```
