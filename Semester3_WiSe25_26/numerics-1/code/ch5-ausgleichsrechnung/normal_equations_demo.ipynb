# Moindres carrés par équations normales

**Référence :** Kröger, §5.2, §5.4.

## Dérivation (Satz 5.2)

Minimiser $\|Ax - b\|_2^2$ → dérivée nulle → **équations normales** :
$$A^T A \, x = A^T b$$

## Problème de stabilité (§5.4)

$$\kappa(A^T A) = \kappa(A)^2$$

Les équations normales perdent le **double** de chiffres significatifs
par rapport à QR. C'est pourquoi on préfère Householder (programme 9).

```python
import numpy as np
import matplotlib.pyplot as plt
from normal_equations import (
    equations_normales, tracer_stabilite, tracer_condition_squared,
)
```

## 1. Régression linéaire

```python
rng = np.random.default_rng(42)
x = np.linspace(0, 5, 30)
y = 2.5*x + 1 + rng.normal(0, 0.5, 30)
A = np.column_stack([np.ones_like(x), x])
c = equations_normales(A, y)
print(f'y ≈ {c[0]:.4f} + {c[1]:.4f} x  (attendu : 1 + 2.5 x)')
```

## 2. $\kappa(A^T A) = \kappa(A)^2$ vérifié

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
tracer_stabilite(ax=axes[0])
tracer_condition_squared(ax=axes[1])
plt.tight_layout()
plt.show()
```

Le graphique de gauche montre que l'erreur des équations normales explose
bien avant celle de QR — exactement le facteur $\kappa$ vs $\kappa^2$.
