## Linéarisation de modèles non-linéaires

| Modèle | Transformation | Problème linéaire |
|---|---|---|
| $y = a e^{bx}$ | $\ln y = \ln a + bx$ | linéaire en $(\ln a, b)$ |
| $y = a x^b$ | $\ln y = \ln a + b \ln x$ | linéaire en $(\ln a, b)$ |
| $y = a/(b+x)$ | $1/y = b/a + x/a$ | linéaire en $(b/a, 1/a)$ |

```python
import numpy as np
import matplotlib.pyplot as plt
from least_squares_fitting import (
    fit_lineaire, fit_polynomial, fit_exponentiel, fit_puissance,
    meilleur_modele, tracer_fits,
)
```

## 1. Données exponentielles : $y = 2 e^{0.8x}$

```python
rng = np.random.default_rng(42)
x = np.linspace(0, 3, 20)
y = 2.0 * np.exp(0.8 * x) + rng.normal(0, 0.5, 20)
y = np.maximum(y, 0.1)

res_lin = fit_lineaire(x, y)
res_exp = fit_exponentiel(x, y)
res_poly = fit_polynomial(x, y, 2)
print(f'Linéaire    : {res_lin}')
print(f'Exponentiel : {res_exp}')
print(f'Quadratique : {res_poly}')
print(f'Meilleur    : {meilleur_modele(x, y)}')
```

```python
tracer_fits(x, y, [res_lin, res_exp, res_poly])
plt.show()
```

## 2. Loi de puissance : $y = 3 x^{0.5}$

```python
x2 = np.linspace(0.5, 10, 25)
y2 = 3.0 * x2**0.5 + rng.normal(0, 0.3, 25)
y2 = np.maximum(y2, 0.1)
res = fit_puissance(x2, y2)
print(res)
tracer_fits(x2, y2, [res, fit_lineaire(x2, y2)])
plt.show()
```
