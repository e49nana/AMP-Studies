# QR-Algorithmus

**Référence :** Kröger, §6.7.

## Principe (§6.7.2)

Itérer :
1. $Q_k R_k = A_k$ (QR-Zerlegung)
2. $A_{k+1} = R_k Q_k$

La suite $A_k$ converge vers une matrice triangulaire supérieure dont les
éléments diagonaux sont les valeurs propres (Schur-Zerlegung, Satz 6.19).

## Accélérations

- **Hessenberg** (§6.7.4) : réduction préalable, chaque pas QR passe de $O(n^3)$ à $O(n^2)$
- **Shift** (§6.7.3) : $\mu = a_{nn}$, convergence cubique pour matrices symétriques

```python
import numpy as np
import matplotlib.pyplot as plt
from qr_algorithm import qr_algorithmus, tracer_convergence_qr
```

```python
A = np.array([[5,1,0,0.5,0],[1,4,1,0,0.3],[0,1,3,1,0],[0.5,0,1,2,0.5],[0,0.3,0,0.5,1]], dtype=float)
print(f'numpy : {np.sort(np.linalg.eigvalsh(A))}')

res_no = qr_algorithmus(A, shift=False, use_hessenberg=False)
res_h  = qr_algorithmus(A, shift=False, use_hessenberg=True)
res_s  = qr_algorithmus(A, shift=True, use_hessenberg=True)

print(f'Sans Hessenberg : {res_no.iterations} it.')
print(f'+ Hessenberg    : {res_h.iterations} it.')
print(f'+ shift         : {res_s.iterations} it.')
```

```python
tracer_convergence_qr(
    [res_no, res_h, res_s],
    ['Basique', 'Hessenberg', 'Hessenberg + shift'])
plt.show()
```

Le QR-Algorithmus est la méthode standard de `numpy.linalg.eigvals`. En pratique,
on utilise la variante **Francis QR** avec double shift implicite et déflation, mais
le principe fondamental est le même.
