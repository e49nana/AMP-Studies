# Vektoriteration (Von-Mises) et Wielandt

**Référence :** Kröger, §6.4.

## Von-Mises (Potenzmethode)

Itérer $y = Ax$, $x \leftarrow y/\|y\|$ converge vers le vecteur propre
dominant. Rate : $|\lambda_2 / \lambda_1|$ (Satz 6.7).

## Wielandt (inverse iteration)

Appliquer Von-Mises à $(A - \sigma I)^{-1}$ : converge vers la valeur
propre **la plus proche de $\sigma$**. Plus $\sigma$ est proche, plus c'est rapide.

```python
import numpy as np
import matplotlib.pyplot as plt
from power_iteration import von_mises, wielandt, tracer_convergence_eigenvalue
```

```python
A = np.array([[4,1,0],[1,3,1],[0,1,2]], dtype=float)
eigvals = np.sort(np.linalg.eigvalsh(A))
print(f'Valeurs propres exactes : {eigvals}')

res_vm = von_mises(A)
print(f'Von-Mises → λ = {res_vm.eigenvalue:.10f} en {res_vm.iterations} it.')

for sigma in [1.5, 3.0, 4.5]:
    res_w = wielandt(A, sigma)
    print(f'Wielandt (σ={sigma}) → λ = {res_w.eigenvalue:.10f} en {res_w.iterations} it.')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
tracer_convergence_eigenvalue([von_mises(A, tol=1e-14)], eigvals[-1], ax=axes[0])
axes[0].set_title('Von-Mises → λ dominant')
tracer_convergence_eigenvalue(
    [wielandt(A, s, tol=1e-14) for s in [1.5, 2.5, 4.0]],
    eigvals[0], ax=axes[1])
axes[1].set_title('Wielandt → λ proche de σ')
plt.tight_layout()
plt.show()
```
