# Méthode de Jacobi pour les valeurs propres

**Référence :** Kröger, §6.5.

## Principe

Itérer des rotations de Givens $A^{(k+1)} = (U^{(k)})^T A^{(k)} U^{(k)}$ (Eq. 6.4)
pour annuler les éléments hors-diagonaux un par un.

Mesure de convergence : $\text{off}(A) = \sqrt{\sum_{i \neq j} a_{ij}^2} \to 0$ (§6.5.7).

Deux variantes :
- **Classique** : annuler le plus grand $|a_{pq}|$
- **Cyclique** : parcourir systématiquement tous les $(p,q)$

```python
import numpy as np
import matplotlib.pyplot as plt
from jacobi_eigenvalue import jacobi_classique, jacobi_cyclique, tracer_convergence_jacobi
```

```python
A = np.array([[4,1,0.5,0],[1,3,0.3,0.1],[0.5,0.3,2,0.2],[0,0.1,0.2,1]], dtype=float)
res_c = jacobi_classique(A)
res_cy = jacobi_cyclique(A)
print(f'numpy       : {np.sort(np.linalg.eigvalsh(A))}')
print(f'Classique   : {np.sort(res_c.eigenvalues)} ({res_c.iterations} rotations)')
print(f'Cyclique    : {np.sort(res_cy.eigenvalues)} ({res_cy.iterations} sweeps)')
```

```python
# Vérification AV = VΛ
Lambda = np.diag(res_c.eigenvalues)
err = np.linalg.norm(A @ res_c.eigenvectors - res_c.eigenvectors @ Lambda, np.inf)
print(f'||AV - VΛ||_∞ = {err:.2e}')
```

```python
tracer_convergence_jacobi([res_c, res_cy], ['Classique', 'Cyclique'])
plt.show()
```
