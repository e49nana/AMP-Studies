# Cercles de Gershgorin

**Référence :** Kröger, §6.6, Satz 6.12.

## Théorème (Satz 6.12)

Chaque valeur propre de $A$ est contenue dans au moins un cercle
$$K_i = \{z \in \mathbb{C} : |z - a_{ii}| \leq r_i\}, \quad r_i = \sum_{k \neq i} |a_{ik}|$$

Par le Korollar 6.14, on peut aussi utiliser les **rayons par colonnes** $\tilde{r}_i = \sum_{k \neq i} |a_{ki}|$.
L'intersection des deux familles donne un encadrement plus serré.

```python
import numpy as np
import matplotlib.pyplot as plt
from gershgorin_circles import tracer_gershgorin, verifier_gershgorin
```

## Beispiel 6.16 (symétrique)

```python
A = np.array([[2,0.4,-0.1],[0.4,3,0.3],[-0.1,0.3,5]], dtype=float)
tracer_gershgorin(A, 'Beispiel 6.16')
plt.show()
for k, v in verifier_gershgorin(A).items():
    print(f'  {k} dans un cercle : {"✓" if v else "✗"}')
```

## Beispiel 6.17 — cercles disjoints

Si un cercle est isolé, il contient **exactement** une valeur propre.

```python
A2 = np.array([[2, 0.7], [0.2, 3]], dtype=float)
tracer_gershgorin(A2, 'Beispiel 6.17', lignes=True, colonnes=True)
plt.show()
```

## Beispiel 6.18 (complexe)

```python
A3 = np.array([[3+1j, 0.3+0.4j], [0.2, 4-0.5j]])
tracer_gershgorin(A3, 'Beispiel 6.18')
plt.show()
```
