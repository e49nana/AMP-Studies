# Interpolation de Lagrange

**Référence :** Kröger, §4.2.2, formules 4.2 et 4.3.

## Fonctions de base $L_i^n(x)$

$$L_i^n(x) = \prod_{j \neq i} \frac{x - x_j}{x_i - x_j}$$

Propriété clé : $L_i(x_j) = \delta_{ij}$ (1 si $i=j$, 0 sinon).

## Polynôme interpolant

$$p(x) = \sum_{i=0}^n y_i \, L_i^n(x)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from lagrange_interpolation import (
    lagrange_interpolation, lagrange_barycentrique,
    tracer_base_lagrange,
)
```

## 1. Visualisation des fonctions de base

```python
xs = np.array([0, 1, 2, 3], dtype=float)
tracer_base_lagrange(xs)
plt.show()
```

Chaque $L_i$ vaut 1 en $x_i$ et 0 aux autres nœuds — c'est ce qui assure $p(x_i) = y_i$.

## 2. Forme barycentrique

La forme standard est $O(n^2)$ par point. La forme barycentrique est $O(n)$ et numériquement plus stable :
$$p(x) = \frac{\sum_i w_i y_i / (x - x_i)}{\sum_i w_i / (x - x_i)}, \quad w_i = \frac{1}{\prod_{j \neq i}(x_i - x_j)}$$

```python
xs = np.array([0, 1, 3, 4], dtype=float)
ys = np.array([1, 2, 0, 5], dtype=float)
x_test = np.linspace(-0.5, 4.5, 200)

p_std = lagrange_interpolation(xs, ys, x_test)
p_bary = lagrange_barycentrique(xs, ys, x_test)
print(f'||standard - barycentrique||_∞ = {np.max(np.abs(p_std - p_bary)):.2e}')

plt.plot(x_test, p_std, 'b-', linewidth=2, label='$p(x)$')
plt.plot(xs, ys, 'ro', markersize=8, label='données')
plt.legend(); plt.grid(True, alpha=0.3); plt.title('Interpolation de Lagrange')
plt.show()
```
