# Différences divisées et forme de Newton

**Référence :** Kröger, §4.2.5, Satz 4.8, Übung 4.5.

## Tableau des différences divisées (formule 4.5)

$$f[x_j, \dots, x_{j+k}] = \frac{f[x_{j+1}, \dots, x_{j+k}] - f[x_j, \dots, x_{j+k-1}]}{x_{j+k} - x_j}$$

## Forme de Newton (Satz 4.8)

$$p(x) = c_0 + c_1(x-x_0) + c_2(x-x_0)(x-x_1) + \cdots$$

avec $c_k = f[x_0, \dots, x_k]$ (première ligne du tableau).

**Évaluation par Horner (formule 4.6) :** $O(n)$ par point, vs $O(n^2)$ pour Lagrange.

```python
import numpy as np
import matplotlib.pyplot as plt
from newton_divided_differences import (
    afficher_tableau, coefficients_newton, evaluer_newton,
)
```

## Übung 4.5

```python
xs = np.array([1, 2, 4], dtype=float)
ys = np.array([6, 6, 0], dtype=float)
afficher_tableau(xs, ys)

c = coefficients_newton(xs, ys)
print(f'\np(x) = {c[0]} + {c[1]}(x-{xs[0]}) + ({c[2]})(x-{xs[0]})(x-{xs[1]})')
print(f'     = -x² + 3x + 4')
```

```python
x_fine = np.linspace(0, 5, 200)
y_fine = evaluer_newton(xs, c, x_fine)
plt.plot(x_fine, y_fine, 'b-', linewidth=2, label='$p(x)$')
plt.plot(xs, ys, 'ro', markersize=8, label='données')
plt.legend(); plt.grid(True, alpha=0.3); plt.title('Übung 4.5')
plt.show()
```

**Avantage de Newton sur Lagrange :** ajouter un point $(x_{n+1}, y_{n+1})$ ne coûte
que $O(n)$ — on calcule juste le nouveau coefficient $c_{n+1} = f[x_0, \dots, x_{n+1}]$
sans recalculer les précédents.
