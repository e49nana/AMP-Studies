# Splines cubiques naturelles

**Référence :** Kröger, §4.3, Définition 4.12.

## 1. Motivation

Le programme 6 a montré que l'interpolation polynomiale de degré élevé
diverge sur nœuds équidistants (Runge). Les **splines** résolvent ce problème :
polynômes de degré 3 par morceaux, raccordés $C^2$.

Sur chaque intervalle $[x_i, x_{i+1}]$ :
$$S_i(x) = \frac{M_i(x_{i+1}-x)^3}{6h_i} + \frac{M_{i+1}(x-x_i)^3}{6h_i} + \left(\frac{y_i}{h_i} - \frac{M_i h_i}{6}\right)(x_{i+1}-x) + \left(\frac{y_{i+1}}{h_i} - \frac{M_{i+1} h_i}{6}\right)(x-x_i)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from splines import SplineCubique, tracer_spline_vs_poly, tracer_erreur_spline
```

## 2. Exemple simple

```python
xs = np.array([0, 1, 2, 3, 4], dtype=float)
ys = np.array([0, 1, 0, 1, 0], dtype=float)
spl = SplineCubique(xs, ys)
print(f'Moments M = {spl.M}')

from scipy.interpolate import CubicSpline
spl_scipy = CubicSpline(xs, ys, bc_type='natural')
x_test = np.linspace(0, 4, 50)
print(f'||mine - scipy||_∞ = {np.max(np.abs(spl(x_test) - spl_scipy(x_test))):.2e}')
```

## 3. Spline vs polynôme sur Runge

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
tracer_spline_vs_poly(n=15, ax=axes[0])
tracer_erreur_spline(ax=axes[1])
plt.tight_layout()
plt.show()
```

Le spline **converge** là où le polynôme **diverge**. C'est la solution standard
en CAO, graphisme, et traitement de données.
