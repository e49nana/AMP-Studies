# Schéma de Neville-Aitken

**Référence :** Kröger, Lemma 4.6, Übung 4.7.

## Formule récursive

$$p_{j,k}(x) = \frac{(x_k - x) \, p_{j,k-1}(x) + (x - x_j) \, p_{j+1,k}(x)}{x_k - x_j}$$

On part de $p_{j,j}(x) = y_j$ et on « monte » le tableau triangulaire.
Le résultat final est $p_{0,n}(x)$.

**Avantage :** pas besoin de construire les coefficients du polynôme —
on obtient directement $p(x)$ en un point.

```python
import numpy as np
import matplotlib.pyplot as plt
from neville_aitken import afficher_tableau, neville_valeur, neville_erreur_estimation
```

## Übung 4.7

```python
xs = np.array([0, 1, 3, 4], dtype=float)
ys = np.array([12, 3, -3, 12], dtype=float)
afficher_tableau(xs, ys, 2.0)
```

```python
print(f'Estimation erreur : {neville_erreur_estimation(xs, ys, 2.0):.6f}')

x_fine = np.linspace(-0.5, 4.5, 200)
y_fine = [neville_valeur(xs, ys, xi) for xi in x_fine]
plt.plot(x_fine, y_fine, 'b-', linewidth=2, label='$p(x)$')
plt.plot(xs, ys, 'ro', markersize=8, label='données')
plt.axhline(0, color='grey', linewidth=0.5)
plt.legend(); plt.grid(True, alpha=0.3); plt.title('Übung 4.7')
plt.show()
```
