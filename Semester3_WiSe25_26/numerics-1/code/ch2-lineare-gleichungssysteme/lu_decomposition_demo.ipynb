## Avantage clé (§2.3.2)

Une fois $PA = LR$ calculée ($\frac{1}{3}n^3$), résoudre $Ax = b$ pour un
**nouveau** $b$ ne coûte que $n^2$ (deux substitutions).

```python
import numpy as np
import matplotlib.pyplot as plt
from lu_decomposition import FactorisationLR, nachiteration, cholesky, tracer_nachiteration
```

## 1. Multiples seconds membres

```python
A = np.array([[2,1,1],[4,3,3],[8,7,9]], dtype=float)
lr = FactorisationLR(A)
for i in range(3):
    b = np.zeros(3); b[i] = 1.0
    print(f'e_{i+1} → x = {lr.resoudre(b)}  (col. {i+1} de A⁻¹)')
print(f'det(A) = {lr.determinant():.1f}')
```

## 2. Nachiteration (§2.3.7)

Améliore itérativement une solution approchée :
$$r = Ax̃ - b, \quad Ay = r, \quad x̃_2 = x̃ - y$$

```python
tracer_nachiteration()
plt.show()
```

## 3. Cholesky ($A = LL^T$, §2.4)

Pour matrices SPD : coût $\frac{1}{6}n^3$ (moitié du Gauss).

```python
B = np.array([[4,2,1],[2,5,3],[1,3,6]], dtype=float)
L = cholesky(B)
print(f'||LLᵀ - B|| = {np.linalg.norm(L@L.T - B, np.inf):.2e}')
```
