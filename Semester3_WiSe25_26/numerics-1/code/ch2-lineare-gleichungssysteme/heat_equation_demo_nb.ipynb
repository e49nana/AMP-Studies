## Problème

Résoudre $-\Delta T = 0$ sur $[0,1]^2$ avec :
- Bord haut : $T = 100$ (source chaude)
- Trois autres bords : $T = 0$

Discrétisation par schéma à 5 points :
$$4 T_{i,j} - T_{i-1,j} - T_{i+1,j} - T_{i,j-1} - T_{i,j+1} = 0$$

Système creux $n^2 \times n^2$ avec **5 entrées non nulles par ligne**.

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from heat_equation_demo import (
    assemblage_laplacien_2d, jacobi_sparse, tracer_solution,
)
from scipy.sparse.linalg import spsolve
```

## 1. Comparaison des solveurs

```python
for n in [10, 30, 50]:
    A, b = assemblage_laplacien_2d(n, bord_haut=100)
    t0 = time.time(); x_sp = spsolve(A, b); t_sp = time.time()-t0
    t0 = time.time(); x_j, k_j = jacobi_sparse(A, b); t_j = time.time()-t0
    print(f'n={n:>2} ({n*n:>4} inconnues) : '
          f'SciPy {t_sp:.4f}s, Jacobi {k_j} it. ({t_j:.2f}s)')
```

## 2. Visualisation de la solution

```python
n = 50
A, b = assemblage_laplacien_2d(n, bord_haut=100)
x = spsolve(A, b)
tracer_solution(x, n, 'Température ($n=50$)')
plt.show()
```

La chaleur se propage du bord haut (100°) vers les bords froids (0°). Les isothermes
sont visibles comme des bandes de couleur horizontales, légèrement courbées aux coins.
