# Analyse d'erreur pour $Ax = b$

**Référence :** Kröger, §2.3.6, Satz 2.17.

## Borne fondamentale

$$\frac{\|\Delta x\|}{\|x\|} \leq \kappa(A) \cdot \frac{\|\Delta b\|}{\|b\|}$$

- $\kappa(A) = 10^k$ → on perd environ $k$ chiffres décimaux
- À $\kappa = 10^{16}$ → toute la précision double est perdue

```python
import numpy as np
import matplotlib.pyplot as plt
from error_analysis_linsys import (
    erreur_directe, erreur_retrograde, chiffres_perdus,
    tracer_erreur_vs_cond, tracer_chiffres_perdus,
)
```

## 1. Règle pratique

```python
for k in [1, 1e4, 1e8, 1e12, 1e16]:
    print(f'κ = {k:.0e} → perte de ~{chiffres_perdus(k)} chiffres sur 16')
```

## 2. Vérification expérimentale

On génère des matrices de conditionnement contrôlé et on vérifie que
l'erreur reste sous la borne $\kappa \cdot \varepsilon_{mach}$.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
tracer_erreur_vs_cond(ax=axes[0])
tracer_chiffres_perdus(ax=axes[1])
plt.tight_layout()
plt.show()
```

Tous les points sont sous la droite rouge — le Satz 2.17 est vérifié.

**Erreur directe vs rétrograde :**
- *Directe* : $\|x_{calc} - x\| / \|x\|$ → ce qu'on veut minimiser
- *Rétrograde* : $\|Ax_{calc} - b\| / \|b\|$ → mesure la qualité du solveur

Un bon solveur (pivot partiel) a une erreur rétrograde $\sim \varepsilon_{mach}$,
mais l'erreur directe est amplifiée par $\kappa(A)$.
