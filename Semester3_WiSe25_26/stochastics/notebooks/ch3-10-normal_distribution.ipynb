## 3. Approximation normale de $B(100, 0.3)$

```python
from math import comb
n, p = 100, 0.3
for k in [25, 30, 35]:
    exact = sum(comb(n,i)*p**i*(1-p)**(n-i) for i in range(k+1))
    approx = binomiale_approx_normale(k, n, p)
    print(f"P(X≤{k}) : exact={exact:.6f}, approx={approx:.6f}")
```
