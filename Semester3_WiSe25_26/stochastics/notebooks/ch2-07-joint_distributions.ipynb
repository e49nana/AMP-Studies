## 1. Deux dés (indépendants)

```python
from joint_distributions import *
import numpy as np
vals = np.arange(1, 7, dtype=float)
P = np.full((6, 6), 1/36)
loi = LoiJointe(vals, vals, P)
print(f"E[X] = {loi.esperance_X():.4f}")
print(f"Cov(X,Y) = {loi.covariance():.6f}")
print(f"Indépendants ? {loi.est_independant()}")
```

## 2. Variables corrélées

```python
x_vals = np.array([1, 2, 3, 4, 5], dtype=float)
P_corr = np.array([
    [0.10, 0.04, 0.02, 0.01, 0.01],
    [0.04, 0.10, 0.04, 0.02, 0.01],
    [0.02, 0.04, 0.10, 0.04, 0.02],
    [0.01, 0.02, 0.04, 0.10, 0.04],
    [0.01, 0.01, 0.02, 0.04, 0.10],
])
P_corr /= P_corr.sum()
loi_c = LoiJointe(x_vals, x_vals, P_corr)
print(f"ρ(X,Y) = {loi_c.correlation():.4f}")
print(f"Var(X+Y) = {loi_c.variance_somme():.4f}")
```

## 3. Visualisation

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(6, 5))
tracer_loi_jointe_heatmap(loi_c, ax=ax)
plt.tight_layout(); plt.show()
```
