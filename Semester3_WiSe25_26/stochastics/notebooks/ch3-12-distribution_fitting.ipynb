## 1. MLE pour la normale

```python
from distribution_fitting import *
import numpy as np
rng = np.random.default_rng(42)
data = rng.normal(5, 2, 500)
mle = mle_normale(data)
print(f"μ̂ = {mle['mu_hat']:.4f} (vrai: 5)")
print(f"σ̂ = {mle['sigma_hat']:.4f} (vrai: 2)")
```

## 2. Test KS

```python
ks = test_ks(data, "norm", loc=mle['mu_hat'], scale=mle['sigma_hat'])
print(f"KS stat = {ks['statistic']:.4f}, p = {ks['p_value']:.4f}")
print(f"Rejet à 5% ? {ks['rejet_5%']}")
```

## 3. Visualisation

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
tracer_ajustement(data, "N(5, 4)", ax=axes[0, 0])
tracer_qq(data, nom="N(5, 2²)", ax=axes[0, 1])
tracer_ecdf_vs_cdf(data, "norm", ax=axes[1, 0], loc=mle['mu_hat'], scale=mle['sigma_hat'])
data_exp = rng.exponential(2, 300)
tracer_qq(data_exp, "expon", nom="Exp", ax=axes[1, 1])
plt.tight_layout(); plt.show()
```
