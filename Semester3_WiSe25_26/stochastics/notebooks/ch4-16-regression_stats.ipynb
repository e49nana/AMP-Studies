## 1. Régression avec inférence
Vrais paramètres : $\beta_0 = 3$, $\beta_1 = 2.5$, $\sigma = 2$

```python
from regression_stats import *
import numpy as np
rng = np.random.default_rng(42)
x = np.linspace(0, 10, 50)
y = 3 + 2.5*x + rng.normal(0, 2, 50)
r = regression_lineaire(x, y)
afficher_regression(r)
```

## 2. Pas de relation ($\beta_1 = 0$)

```python
y_null = 5 + rng.normal(0, 3, 50)
r_null = regression_lineaire(x, y_null)
afficher_regression(r_null)
```

## 3. Visualisation avec bandes de confiance

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
tracer_regression(x, y, "(signal)", ax=axes[0])
tracer_regression(x, y_null, "(bruit pur)", ax=axes[1])
plt.tight_layout(); plt.show()
```
