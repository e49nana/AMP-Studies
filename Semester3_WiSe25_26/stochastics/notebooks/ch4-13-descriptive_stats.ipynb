## 2. Vérification from-scratch vs NumPy

```python
print(f"moyenne : {moyenne(data):.6f} vs {np.mean(data):.6f}")
print(f"médiane : {mediane(data):.6f} vs {np.median(data):.6f}")
print(f"var(s²) : {variance_echantillon(data):.4f} vs {np.var(data, ddof=1):.4f}")
```
