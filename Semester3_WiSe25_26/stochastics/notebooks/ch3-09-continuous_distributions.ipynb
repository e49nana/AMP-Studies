## 2. Normale — règle 68-95-99.7

```python
from scipy import stats
for k in [1, 2, 3]:
    p = stats.norm.cdf(k) - stats.norm.cdf(-k)
    print(f"P(|Z| < {k}) = {p:.4f} = {p*100:.1f}%")
```
