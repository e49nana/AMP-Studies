## 1. TCL pour différentes distributions
$(\bar{X}_n - \mu) / (\sigma/\sqrt{n}) \to N(0,1)$

```python
from central_limit_theorem import *
for dist, params, nom in [("uniforme", {}, "U(0,1)"), ("exponentielle", {"lam": 1}, "Exp(1)"),
                            ("bernoulli", {"p": 0.3}, "Bern(0.3)"), ("de", {}, "Dé")]:
    print(f"{nom:15s} :", end="")
    for n in [1, 5, 30]:
        z = moyennes_echantillon(dist, n, **params)
        ks = test_normalite_ks(z)
        print(f"  n={n:>2} KS={ks['KS_stat']:.4f}{'✓' if ks['normal'] else '✗'}", end="")
    print()
```

## 2. Application : sondage

```python
import numpy as np
p, n = 0.6, 1000
sigma = np.sqrt(p*(1-p)/n)
print(f"IC 95% : [{p-1.96*sigma:.4f}, {p+1.96*sigma:.4f}]")
print(f"Marge : ±{1.96*sigma*100:.1f}%")
```
