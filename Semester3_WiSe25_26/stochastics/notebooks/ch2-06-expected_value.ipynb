## 2. Inégalité de Tchebychev
$P(|X-\mu| \geq k\sigma) \leq 1/k^2$

```python
rng = np.random.default_rng(42)
X = rng.normal(0, 1, 100_000)
for k in [1, 2, 3]:
    print(f"k={k} : borne = {tchebychev(k):.4f}, réel = {np.mean(np.abs(X) >= k):.4f}")
```
